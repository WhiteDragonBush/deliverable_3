```python
# TASK 2
import pandas
from matplotlib import pyplot
import numpy

df = pandas.read_csv("Pokemon.csv")
df
```

```python
# TASK 3
legendary = df[df['Legendary'] == True]
nonlegendary = df[df['Legendary'] == False]

pyplot.tight_layout()
fig = pyplot.figure(figsize=(15,3))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

weights1 = numpy.ones_like(nonlegendary['Speed']) / len(nonlegendary['Speed'])
weights2 = numpy.ones_like(legendary['Speed']) / len(legendary['Speed'])
ax1.hist(nonlegendary['Speed'], edgecolor='blue', fill=False, label='non-legendary', weights=weights1)
ax1.hist(legendary['Speed'], edgecolor='orange', fill=False, label='legendary', weights=weights2)
ax1.set_xlabel("Speed")
ax1.set_ylabel("Density")
ax1.legend()

ax2.scatter(nonlegendary['Defense'], nonlegendary['HP'], c='blue', label='non-legendary', alpha=0.5)
ax2.scatter(legendary['Defense'], legendary['HP'], c='orange', label='legendary', alpha=0.5)
ax2.set_xlabel("Defense")
ax2.set_ylabel("HP")
ax2.legend()

ax3.scatter(nonlegendary['Sp. Atk'], nonlegendary['Attack'], c='blue', label='non-legendary', alpha=0.5)
ax3.scatter(legendary['Sp. Atk'], legendary['Attack'], c='orange', label='legendary', alpha=0.5)
ax3.set_xlabel("Sp. Atk")
ax3.set_ylabel("Attack")
ax3.legend()
```
